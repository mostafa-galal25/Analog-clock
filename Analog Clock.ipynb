#   Analog Clock

```python
import cv2
import numpy as np
import math
import time
import datetime


width, height = 800, 800
center = (width /2, height /2)
big_rad = 100

image = np.zeros((800, 800, 3), dtype=np.uint8)


while True:
    image.fill(0)
    now = datetime.datetime.now()

    sec = math.radians(((now.second + now.microsecond/1000000) * 6)-90)
    minute_angle = (now.minute / 60) * 2 * math.pi - math.pi / 2
    hour_angle = ((now.hour % 12) / 12) * 2 * math.pi - math.pi / 2


    #sec_angle 
    xSec = 400 + int(100 * math.cos(sec))
    ySec = 400 + int(100 * math.sin(sec))
    
    #min_angle 
    x2 = int(center[0] + 70 * math.cos(minute_angle))
    y2 = int(center[1] + 70 * math.sin(minute_angle))
    
    #h_angle 
    x3 = int(center[0] + 60 * math.cos(hour_angle))
    y3 = int(center[1] + 60 * math.sin(hour_angle))

    cv2.circle(image, (400,400), big_rad, (50, 50, 50), 2)
    
    cv2.line(image, (400,400), (xSec,ySec), (0, 0, 255), 1)
    cv2.line(image, (400,400), (x2,y2), (0, 255, 255), 2)
    cv2.line(image, (400,400), (x3,y3), (255, 150,150), 3)
    
    # Draw hour labels
    for i in range(1, 13):
        angle = (i / 12) * 2 * math.pi - math.pi / 2
        label_x = int(center[0] + (big_rad+15 ) * math.cos(angle))
        label_y = int(center[1] + (big_rad+15 ) * math.sin(angle))
        cv2.putText(image, str(i), (label_x-15, label_y+10), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255), 3, cv2.LINE_AA)

    date_text = now.strftime("%Y-%m-%d")
    time_text=now.strftime("%H:%M:%S")
    cv2.putText(image, "Date: " +date_text, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255), 2, cv2.LINE_AA)
    cv2.putText(image, "Time: " +time_text, (10, 60), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255), 2, cv2.LINE_AA)
        
    cv2.imshow("clock", image)

    
    k=cv2.waitKey(1)
    if k==ord('q'):
        break
cv2.destroyAllWindows()
```
